# Задача 3

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype(str), showfliers=False);
```

Вывод: Стоимость недвижимости зависит от вида. Дома с видом на набережную стоят дороже.

```python
sns.boxplot(x=df['price'], y=df['view'].astype(str), showfliers=False);
```

Вывод: Чем лучше вид, тем выше цена недвижимости

```python
sns.boxplot(x=df['price'], y=df['grade'].astype(str), showfliers=False);
```

Вывод: Чем современней дизайн интерьера недвижимости и качественнейтматериалы, тем выше стоимость недвижимости

```python
corr_matrix= df.corr()
corr_matrix=np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix)<0.3]=0
corr_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, linewidths=1, cmap='Spectral');
```

Вывод: 
Наибольшое влияение на стоимость объекта недвижимости оказывает:
- Жилая площадь (0.7),
- Материалы и дизайн (0.7),
- площадь мансарды (0,6)
- пжилая площадь соседей (0,6)
- количество ванных комнат (0,5)
